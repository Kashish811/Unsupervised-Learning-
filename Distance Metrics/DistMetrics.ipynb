**Method 1:**

Load images (grayscale)

```python
from PIL import Image
import numpy as np

img1 = Image.open("/content/OAS1_0028_MR1_mpr-1_100.jpg").convert("L")
img2 = Image.open("/content/OAS1_0028_MR1_mpr-1_101.jpg").convert("L")
```

Resize to a fixed size

```python
SIZE = (128, 128)

img1 = img1.resize(SIZE)
img2 = img2.resize(SIZE)
```

Convert to NumPy arrays


```python
arr1 = np.array(img1, dtype=np.float32)
arr2 = np.array(img2, dtype=np.float32)
```

Normalize pixel values


```python
arr1 /= 255.0
arr2 /= 255.0
```

Flatten images into vectors


```python
vec1 = arr1.flatten()
vec2 = arr2.flatten()
```

Euclidean Distance


```python
euclidean = np.linalg.norm(vec1 - vec2)
print("Euclidean Distance:", euclidean)
```

Manhattan Distance

```python
manhattan = np.sum(np.abs(vec1 - vec2))
print("Manhattan Distance:", manhattan)
```

Chebyshev Distance


```python
# Chebyshev Distance (L-infinity)
chebyshev = np.max(np.abs(vec1 - vec2))
print("Chebyshev Distance:", chebyshev)
```

Minkowskii with p=1


```python
# Minkowski Distance with p = 1
p = 1
minkowski_p1 = np.power(np.sum(np.abs(vec1 - vec2) ** p), 1 / p)
print("Minkowski Distance (p=1):", minkowski_p1)
```

Minkowski Distance (p = 2) → Euclidean Distance


```python
# Minkowski Distance with p = 2
p = 2
minkowski_p2 = np.power(np.sum(np.abs(vec1 - vec2) ** p), 1 / p)
print("Minkowski Distance (p=2):", minkowski_p2)
```

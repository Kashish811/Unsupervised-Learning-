**CASE 2: Logistic Regression WITH PCA**

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/content/breast-cancer.csv")

#Drop ID column
df.drop(columns=['id'], inplace=True)

#Encode target variable
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
#Split Features and Target
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
```

```python
#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Feature Scaling
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**PCA CASE 1: PC1 only**

```python
pca_1 = PCA(n_components=1)

X_train_pca1 = pca_1.fit_transform(X_train_scaled)
X_test_pca1 = pca_1.transform(X_test_scaled)

model_1 = LogisticRegression(max_iter=5000)
model_1.fit(X_train_pca1, y_train)

y_pred_1 = model_1.predict(X_test_pca1)
acc_1 = accuracy_score(y_test, y_pred_1)

explained_var = pca_1.explained_variance_ratio_
print("Variance: ", explained_var)

print("Accuracy using PC1:", acc_1)
print("Model Report")
print(classification_report(y_test, y_pred_1, zero_division=1))
```

**PCA CASE 2: PCA1 + PCA2**

```python
pca_2 = PCA(n_components=2)

X_train_pca2 = pca_2.fit_transform(X_train_scaled)
X_test_pca2 = pca_2.transform(X_test_scaled)

explained_var = pca_2.explained_variance_ratio_
print("Variance: ", explained_var)

model_2 = LogisticRegression(max_iter=5000)
model_2.fit(X_train_pca2, y_train)

y_pred_2 = model_2.predict(X_test_pca2)
acc_2 = accuracy_score(y_test, y_pred_2)

print("Accuracy using PC1 + PC2:", acc_2)
print("Model report")
print(classification_report(y_test, y_pred_2, zero_division=1))
```

**PCA CASE 3: PCA1 + PCA2 + PCA3**

```python
pca_3 = PCA(n_components=3)

X_train_pca3 = pca_3.fit_transform(X_train_scaled)
X_test_pca3 = pca_3.transform(X_test_scaled)

explained_var = pca_3.explained_variance_ratio_
print("Variance: ", explained_var)

model_3 = LogisticRegression(max_iter=5000)
model_3.fit(X_train_pca3, y_train)

y_pred_3 = model_3.predict(X_test_pca3)
acc_3 = accuracy_score(y_test, y_pred_3)

print("Accuracy using PC1 + PC2 + PC3:", acc_3)
print("Model Report")
print(classification_report(y_test, y_pred_3, zero_division=1))
```
